# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data)
print(school_count)
```

```python
# Calculate the total number of students
student_count = len(student_data)
print(student_count)
```

```python
# Calculate the total budget

total_budget = school_data["budget"].sum()
print(total_budget)
```

```python
# Calculate the average (mean) math score
average_math_score = student_data["math_score"].mean()
print(average_math_score)
```

```python
# Calculate the average (mean) reading score
average_reading_score = student_data["reading_score"].mean()
print(average_reading_score)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame(
    {"Total Number of Unique Schools: " :[school_count],
    "Total Students: " : [student_count],
    "Total Budget: ": [total_budget],
    "Average Math Score: ": [average_math_score],
    "Average Reading Score: " : [average_reading_score],
    "Percentage of Student Body Passing Math: " : [passing_math_percentage],
    "Percentage of Student Body Passing Reading: " : [passing_reading_percentage],
    "Percentage of Student Body Passing Both Math & Reading: " : [overall_passing_rate]
    }
)

# Formatting
# district_summary["Total Number of Unique Schools: "] = district_summary["Total Number of Unique Schools: "].map("{:,}".format)
district_summary["Total Students: "] = district_summary["Total Students: "].map("{:,}".format)
district_summary["Total Budget: "] = district_summary["Total Budget: "].map("${:,.2f}".format)
district_summary["Average Math Score: "] = district_summary["Average Math Score: "].map("{:,.0f}".format)
district_summary["Average Reading Score: "] = district_summary["Average Reading Score: "].map("{:,.0f}".format)
district_summary["Percentage of Student Body Passing Math: "] = district_summary["Percentage of Student Body Passing Math: "].map("{:.2f}%".format)
district_summary["Percentage of Student Body Passing Reading: "] = district_summary["Percentage of Student Body Passing Reading: "].map("{:.2f}%".format)
district_summary["Percentage of Student Body Passing Both Math & Reading: "] = district_summary["Percentage of Student Body Passing Both Math & Reading: "].map("{:.2f}%".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count
per_school_counts = school_data_complete.groupby(["school_name"]).count()["student_name"]
print(per_school_counts)
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
print(per_school_capita)
print(per_school_budget)   
```

```python
# Calculate the average test scores
per_school_math = school_data_complete.groupby(["school_name"]).agg(math_score = ('math_score', 'mean'))
per_school_reading = school_data_complete.groupby(["school_name"]).agg(reading_score = ('reading_score', 'mean'))
print(per_school_math)
print(per_school_reading)
```

```python
# Calculate the number of students (originally said 'schools' here instead of 'students') 
# with math scores of 70 or higher

student_passing_math = school_data_complete[(school_data_complete['math_score']>=70)]
#student_passing_math = student_passing_math.groupby(["school_name"]).count()[['math_score']]

print(student_passing_math)

# Calculate the number of schools with math scores of 70 or higher

school_passing_math = per_school_math.loc[(per_school_math['math_score']>=70)]
print(school_passing_math)
school_passing_math.count()
```

```python
# Calculate the number of students (originally said 'schools' here instead of 'students') 
# with math scores of 70 or higher

student_passing_reading = school_data_complete[(school_data_complete['reading_score']>=70)]
#student_passing_reading = student_passing_reading.groupby(["school_name"]).count()[['reading_score']]

print(student_passing_reading)

# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = per_school_reading.loc[(per_school_reading['reading_score'] >= 70)]
print(school_passing_reading)
school_passing_reading.count()
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
passing_math_and_reading['student_name'].count()

```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = student_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = student_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
```

```python
per_school_math
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({'school_types':school_types,
'Total Students':per_school_counts,
'Total School Budget':per_school_budget,
'Per Student Budget':per_school_capita,
'Average Math Score':per_school_math['math_score'],
'Average Reading Score':per_school_reading['reading_score'],
'Percentage of Student Body Passing Math':per_school_passing_math,
'Percentage of Student Body Passing Reading':per_school_passing_reading,
'Percentage of Student Body Passing Both Math and Reading':overall_passing_rate})

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
# per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:.1f}".format)
# per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:.1f}".format)
# per_school_summary["Percentage of Student Body Passing Math"] = per_school_summary["Percentage of Student Body Passing Math"].map("{:.1f}%".format)
# per_school_summary["Percentage of Student Body Passing Reading"] = per_school_summary["Percentage of Student Body Passing Reading"].map("{:.1f}%".format)
# per_school_summary["Percentage of Student Body Passing Both Math and Reading"] = per_school_summary["Percentage of Student Body Passing Both Math and Reading"].map("{:.1f}%".format)

# Sort by school_types
per_school_summary = per_school_summary.sort_values('school_types')

# Display the DataFrame
per_school_summary
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
ninth_graders_scores = ninth_graders.groupby(['school_name']).mean()
tenth_graders_scores = tenth_graders.groupby(['school_name']).mean()
eleventh_graders_scores = eleventh_graders.groupby(['school_name']).mean()
twelfth_graders_scores = twelfth_graders.groupby(['school_name']).mean()


# Use the code to select only the `math_score`.
ninth_grade_math_scores = ninth_graders_scores[["math_score"]]
tenth_grade_math_scores = tenth_graders_scores[["math_score"]]
eleventh_grade_math_scores = eleventh_graders_scores[["math_score"]]
twelfth_grade_math_scores = twelfth_graders_scores[["math_score"]]

# Format math score data
ninth_grade_math_scores = ninth_graders_scores['math_score'].map("{:.1f}".format)
tenth_grade_math_scores = tenth_graders_scores['math_score'].map("{:.1f}".format)
eleventh_grade_math_scores = eleventh_graders_scores['math_score'].map("{:.1f}".format)
twelfth_grade_math_scores = twelfth_graders_scores['math_score'].map("{:.1f}".format)


# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.concat([ninth_grade_math_scores,
tenth_grade_math_scores,
eleventh_grade_math_scores,
twelfth_grade_math_scores], axis=1)

# Change headers
math_scores_by_grade.columns=['9th','10th','11th','12th']

#Minor data wrangling
math_scores_by_grade.index.name = None

#Display the DataFrame
math_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df.dtypes
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita,
bins = spending_bins, labels=labels)
```

```python
# school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()
school_spending_df.dtypes
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()['Average Math Score']
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Percentage of Student Body Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Percentage of Student Body Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Percentage of Student Body Passing Both Math and Reading"]
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({'Average Math Score':spending_math_scores,
'Average Reading Score':spending_reading_scores,
'Percentage of Student Body Passing Math':spending_passing_math,
'Percentage of Student Body Passing Reading':spending_passing_reading,
'Percentage of Student Body Passing Both Math and Reading':overall_passing_spending})

# Display results
spending_summary
```
